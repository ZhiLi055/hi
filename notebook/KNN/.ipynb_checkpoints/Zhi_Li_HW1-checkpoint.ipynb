```python
#load data from dat file and transform to pandas dataframe
dict_2C = {'pelvic incidence':[], 'pelvic tilt':[], 'lumbar lordosis angle':[], 'sacral slope':[], 'pelvic radius':[], 
           'grade of spondylolisthesis':[],'class':[] }

with open("column_2C.dat") as datFile:
    for data in datFile:
        data = data.split()
        dict_2C['pelvic incidence'].append(float(data[0]))
        dict_2C['pelvic tilt'].append(float(data[1]))
        dict_2C['lumbar lordosis angle'].append(float(data[2]))
        dict_2C['sacral slope'].append(float(data[3]))
        dict_2C['pelvic radius'].append(float(data[4]))
        dict_2C['grade of spondylolisthesis'].append(float(data[5]))
        if data[6] == 'NO':
            dict_2C['class'].append(0)
        else:
            dict_2C['class'].append(1)

data_2C = pd.DataFrame(dict_2C, columns=['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle', 'sacral slope', 'pelvic radius', 
                                         'grade of spondylolisthesis','class' ])
```

## $For \ Question \ 1(b)\ iii$

```python
training_set = data_2C[data_2C['class']==0].head(70).copy()
training_set = training_set.append(data_2C[data_2C['class']==1].head(140),ignore_index=True)
test_set = data_2C[data_2C['class']==0][70:].copy()
test_set = test_set.append(data_2C[data_2C['class']==1][140:],ignore_index=True)
print('The first 5 rows of training set are as follows')
display(training_set.head(5))
print('The first 5 rows of test set are as follows')
display(test_set.head(5))
```

## $For \ Question \ 1(c)\ i$

```python
def knn(k,training,test):
    
    if list(training.columns) != list(test.columns):
        print('ERROR! Columns format of training set and test set are different')
        return 
    training_array = training.iloc[:,:-1].values
    test_array = test.iloc[:,:-1].values
    label_sum = []
    for point in test_array:
        dist = np.sum((training_array - point)**2,axis=1)
        order = np.argsort(dist)
        result = []
        for row in range(k):
            result.append(training.iloc[order[row],-1])
        if sum(result) > len(result)/2:
            label = 1
        else:
            label = 0
        label_sum.append(label)
    test_result = test.copy()
    test_result['estimated_class'] = label_sum
    return test_result
```

## $For \ Question \ 1(c)\ ii$

```python
training_error_sum = []
test_error_sum = []
for k in range(208,0,-3):
    #First we used the code in question (c)i to calculate the estimated class for training set and test set respectively
    training_result = knn(k, training_set, training_set)
    test_result = knn(k, training_set, test_set)
    #Then we calculate the corresponding train and test errors
    training_error = training_result[training_result['class'] != training_result['estimated_class']].shape[0]/training_result.shape[0]
    test_error = test_result[test_result['class'] != test_result['estimated_class']].shape[0]/test_result.shape[0]
    training_error_sum.append(training_error)
    test_error_sum.append(test_error)
```

```python
#Plot train and test errors from 208 to 1
sns.set(style='darkgrid')
x = list(range(208,0,-3))
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, training_error_sum, label='train error')
ax.plot(x, test_error_sum,label='test error')
ax.legend()
ax.set_title('Train and test errors for k from 208 to 1', fontsize=16, weight='bold')
ax.set_xlabel('k values',fontsize=15)
ax.set_ylabel('Error rate',fontsize=15)
ax.annotate('optimal k : (4, 0.06)', xy=(4, 0.06), xytext=(45, 0.01),
            arrowprops={'arrowstyle':'->'})
ax.invert_xaxis()
```

#### From the plot we can find that the lowest point of test error is at k = 4 and the corresponding error rate is 0.06, therefore the optimal k is 4 .

```python
#Then we calculate the confusion matrix, true positive rate, true negative rate, precision and F1-score when k = 4.
test_result = knn(4, training_set, test_set)
#For confusion matrix
#We first calculate the number of those who are 0 and predicted 0
actual0_predict0 = test_result[(test_result['class']==0) & (test_result['estimated_class']==0)].shape[0]
actual0_predict1 = test_result[(test_result['class']==0) & (test_result['estimated_class']==1)].shape[0]
actual1_predict1 = test_result[(test_result['class']==1) & (test_result['estimated_class']==1)].shape[0]
actual1_predict0 = test_result[(test_result['class']==1) & (test_result['estimated_class']==0)].shape[0]
confusion_matrix = np.array([[actual0_predict0, actual0_predict1],[actual1_predict0, actual1_predict1]])
#Now we plot the confusion  matrix
ax = sns.heatmap(confusion_matrix, annot=True, cmap='Blues',linewidth=0.5, cbar=False,annot_kws={'size': 20})
ax.set_title('Confusion Matrix',fontsize=20, weight='bold')
ax.set_xlabel('\nPredicted Class',fontsize=15, weight='bold')
ax.set_ylabel('Actual Class ',fontsize=15, weight='bold')
ax.xaxis.set_ticklabels(['0','1'],fontsize=15)
ax.yaxis.set_ticklabels(['0','1'],fontsize=15)
sns.set(rc = {'figure.figsize':(8,6)})
```

#### Then the confusion matrix is as above .

```python
#Now we calculate true positive rate, true negative rate, precision,and F1-score accrodingly.
true_positive_rate = actual1_predict1/(actual1_predict1+actual1_predict0)
true_negative_rate = actual0_predict0/(actual0_predict0+actual0_predict1)
precision = actual1_predict1/(actual1_predict1+actual0_predict1)
F1_score = 2*precision*true_positive_rate/(true_positive_rate+precision)
print('True positive rate is', round(true_positive_rate,2))
print('True negative rate is', round(true_negative_rate,2))
print('precision is', round(precision,2))
print('F1-score is', round(F1_score,2))
```

#### Therefore when k take the optimal value 4, the true positive rate, true negative rate, precision,and F1-score accrodingly are as above

## $For \ Question \ 1(c)\ iii$

```python
#To reduce running time, here we use scikit-learn package to run the KNN algorithms
#First we prepare test data to fit the sciket-learn package
test_result_package = test_set.copy()
best_error = []
test = test_set.iloc[:,:-1].values
#Then we prepare the training data according to N 
for N in range(10,211,10):
    test_error_sum = []
    train_N = training_set[training_set['class']==0].head(round(N/3))
    train_N = train_N.append(training_set[training_set['class']==1].head(round(2*N/3)),ignore_index=True)
    X = train_N.iloc[:,:-1].values
    y = train_N.iloc[:,-1].values
    #For each N calculate test error for each k from 1, 6,...until the number that is just smaller than N.
    for k in range(1,N+1,5):
        neigh = KNeighborsClassifier(n_neighbors=k)
        neigh.fit(X, y)
        test_result_package['package_class'] = neigh.predict(test)
        test_error = test_result_package[test_result_package['class'] != test_result_package['package_class']].shape[0]/test_result_package.shape[0]
        test_error_sum.append(test_error)
    best_error.append(min(test_error_sum))
```

```python
#Plot the learning curve
sns.set(style='darkgrid')
x = list(range(10,211,10))
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, best_error,label='best test error')
ax.set_title('Learning curves for N from 10 to 210', fontsize=16, weight='bold')
ax.set_xlabel('N values',fontsize=15)
a = ax.set_ylabel('Best test error rate',fontsize=15)
```

## $For \ Question \ 1(d)\ i.A$

```python
#First we select the best k for manhattan distance
df_test_error = []
df_optimal_k = []
test_result_package = test_set.copy()
test_error_sum = []
X = training_set.iloc[:,:-1].values
y = training_set.iloc[:,-1].values

for k in range(1,197,5):
    neigh = KNeighborsClassifier(n_neighbors=k,metric='manhattan')
    neigh.fit(X, y)
    test_result_package['package_class'] = neigh.predict(test_set.iloc[:,:-1].values)
    test_error = test_result_package[test_result_package['class'] != test_result_package['package_class']].shape[0]/test_result_package.shape[0]
    test_error_sum.append(test_error)

best_test_error = min(test_error_sum)
# best_k = list(range(1,197,5))[test_error_sum.index(best_test_error)]
test_error_sum.reverse()
best_k = list(range(1,197,5))[(len(test_error_sum)-1)-test_error_sum.index(best_test_error)]
print('The optimal k for manhattan distance is ',best_k, ', with corresponding best test error', best_test_error)
#save the corresponding test error and k value for generating the summarized table later
df_optimal_k.append(best_k)
df_test_error.append(best_test_error)
```

## $For \ Question \ 1(d)\ i.B$

```python
#For this problem we use the optimal k = 6
#First we select the best p for in {10^0.1, 10^0.2, 10^0.3,... 10}
test_result_package = test_set.copy()
test_error_sum = []
X = training_set.iloc[:,:-1].values
y = training_set.iloc[:,-1].values
p_range = [10**i for i in np.linspace(0.1,1,10)]

for p_value in p_range:
    neigh = KNeighborsClassifier(n_neighbors=6,metric='minkowski',p=p_value)
    neigh.fit(X, y)
    test_result_package['package_class'] = neigh.predict(test_set.iloc[:,:-1].values)
    test_error = test_result_package[test_result_package['class'] != test_result_package['package_class']].shape[0]/test_result_package.shape[0]
    test_error_sum.append(test_error)

best_test_error = min(test_error_sum)
best_p = p_range[test_error_sum.index(best_test_error)]
best_logp = math.log10(best_p)
print('The optimal log10(p) for minkowski distance is ',round(best_logp,1), ', with corresponding best test error when k = 6 is', best_test_error)
#save the corresponding test error and k value for different p to generate the summarized table later
df_optimal_k.extend([6] * 10)
df_test_error.extend(test_error_sum)
```

## $For \ Question \ 1(d)\ ii$

```python
#First we select the best k for mahalanobis distance
test_result_package = test_set.copy()
test_error_sum = []
X = training_set.iloc[:,:-1].values
y = training_set.iloc[:,-1].values

for k in range(1,197,5):
    neigh = KNeighborsClassifier(n_neighbors=k,metric='mahalanobis',metric_params={'V': np.cov(X.T)})
    neigh.fit(X, y)
    test_result_package['package_class'] = neigh.predict(test_set.iloc[:,:-1].values)
    test_error = test_result_package[test_result_package['class'] != test_result_package['package_class']].shape[0]/test_result_package.shape[0]
    test_error_sum.append(test_error)

best_test_error = min(test_error_sum)
test_error_sum.reverse()
best_k = list(range(1,197,5))[(len(test_error_sum)-1)-test_error_sum.index(best_test_error)]
print('The optimal k for mahalanobis distance is ',best_k, ', with corresponding best test error', best_test_error)
#save the corresponding test error and k value for generating the summarized table later
df_optimal_k.append(best_k)
df_test_error.append(best_test_error)
```

```python
#Then we can plot the final summarized table
summarized_table = pd.DataFrame({'Test errors':df_test_error, 'Corresponding optimal k':df_optimal_k},
                                columns=['Test errors','Corresponding optimal k'],index=['Manhattan Distance']
                                +['Minkowski Distance when log10(p) is log10({})'.format(round(p,1)) for p in np.linspace(0.1,1,10)]
                                +['Chebyshev Distance']+['Mahalanobis Distance'])
summarized_table
```
